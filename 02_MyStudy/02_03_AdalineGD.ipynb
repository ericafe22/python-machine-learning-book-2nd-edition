```python
# IRISデータセットの読み込み

from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
iris = load_iris()

iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_label = pd.Series(data=iris.target)

X = iris_data.iloc[0:100, [0,2]].values 
y_tmp = iris_label.iloc[0:100,].values
y = np.where(y_tmp == 0, -1 , 1)
```

```python
class AdalineGD(object):
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        print(self.W_)
        self.cost_ = []

        for i in range(self.n_iter):
            net_input = self.net_input(X)
            output = self.activation(net_input)
            print("output")
            print(output)
            print("y")
            print(y)
            errors = (y - output)
            print("errors")
            print(errors)
            self.w_[1:] += self.eta * X.T.dot(errors)  # ★各X値と誤差行列の内積値
            print("X.T.dot(errors)")
            print(X.T.dot(errors))
            print("self.eta * X.T.dot(errors)")
            print(self.eta * X.T.dot(errors))
            print("self.w_[1:]")
            print(self.w_[1:])
            self.w_[0] += self.eta * errors.sum()
            print("errors.sum()")
            print(errors.sum())
            print("self.w_[0]")
            print(self.w_[0])
            cost = (errors**2).sum() / 2.0
            self.cost_.append(cost)
        return self

    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        return X # 恒等関数

    def predict(self, X):
        return np.where(self.activation(self.net_input(X)) >= 0.0, 1, -1)
```
